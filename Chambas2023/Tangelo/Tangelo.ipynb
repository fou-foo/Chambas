```python
numerics_c = ["MONTHS_BALANCE", "CNT_CHILDREN", "AMT_INCOME_TOTAL", "DAYS_BIRTH", "DAYS_EMPLOYED", "CNT_FAM_MEMBERS"]
non_numerics = ['CODE_GENDER', 'FLAG_OWN_CAR',  'FLAG_OWN_REALTY', 'NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS',
                'NAME_HOUSING_TYPE',  'FLAG_MOBIL','FLAG_WORK_PHONE', "FLAG_PHONE" 'FLAG_EMAIL']#, 'OCCUPATION_TYPE'] FLAG_PHONE

```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder

scaler = StandardScaler()
numerics_c = ["MONTHS_BALANCE", "CNT_CHILDREN", "AMT_INCOME_TOTAL", "DAYS_BIRTH", "DAYS_EMPLOYED", "CNT_FAM_MEMBERS"]
non_numerics = ['CODE_GENDER', 'FLAG_OWN_CAR',  'FLAG_OWN_REALTY', 'NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS',
                'NAME_HOUSING_TYPE', 'FLAG_WORK_PHONE',  'FLAG_EMAIL', "FLAG_PHONE"]#, 'OCCUPATION_TYPE'] FLAG_PHONE
data.CODE_GENDER = data.CODE_GENDER.astype(str) 
data.FLAG_OWN_CAR = data.FLAG_OWN_CAR.astype(str) 
data.FLAG_OWN_REALTY = data.FLAG_OWN_REALTY.astype(str) 
data.NAME_INCOME_TYPE = data.NAME_INCOME_TYPE.astype(str) 
data.NAME_EDUCATION_TYPE = data.NAME_EDUCATION_TYPE.astype(str) 
data.NAME_FAMILY_STATUS = data.NAME_FAMILY_STATUS.astype(str) 
data.NAME_HOUSING_TYPE = data.NAME_HOUSING_TYPE.astype(str ) 

data.FLAG_WORK_PHONE = data.FLAG_WORK_PHONE.astype(str) 
data.FLAG_PHONE = data.FLAG_PHONE.astype(str) 
data.FLAG_EMAIL = data.FLAG_EMAIL.astype(str) 
#data.OCCUPATION_TYPE = data.OCCUPATION_TYPE.astype(str) 



ordinal = OneHotEncoder(sparse_output=False)
import numpy as np 
data[numerics_c]  = scaler.fit_transform(data[numerics_c])
```

```python
data[non_numerics].shape
```

```python
for i in non_numerics: 
    print(i)
    print(set(data[i].values))
```

```python
ohe = ordinal.fit(data[non_numerics])
ohe_data = ohe.transform(data[non_numerics])
ohe_data = pd.DataFrame(ohe_data)
ohe_data.columns = ohe.get_feature_names_out()
data_alg = pd.concat( [data[numerics_c], ohe_data], axis=1) 
data_alg
```

```python
data_alg_over, y_over  = oversample.fit_resample( X= data_alg , y  = data["STATUS"]) 
```

```python
y_over.value_counts()
```

```python
p = .8 
seed = 0 

from sklearn.model_selection import train_test_split

train_x, test_x, train_y , test_y = train_test_split( data_alg_over, y_over, train_size=p, random_state=seed)
```

```python
%%time
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import  RocCurveDisplay, classification_report
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier

classifiers = [
    #LogisticRegression(),
    RandomForestClassifier()
    
]
   

for clf in classifiers: 
    # train
    clf.fit(train_x, train_y)
    y_hat = clf.predict(train_x)
    print(classification_report(train_y, y_hat))
    #print(RocCurveDisplay.from_estimator(clf, train_x, train_y))
    plt.show()
 
    # testing 
    y_hat = clf.predict(test_x )
    print(classification_report(test_y, y_hat))
    #RocCurveDisplay.from_estimator(clf, test_y, y_hat)
    plt.show()
    
        
```
