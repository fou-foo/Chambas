## Preprocesamiento de los datos 

Como parte del preprocesamiento de datos realizamos el casteo necesario para la manipulación de las series de tiempo, luego rellenamos los datos nulos por medio de repetir la observación anterior observada. En cuanto a las observaciones repetidas en el mismo día se reducen considerando la media de sus valores. 

No se realizan transformaciones en las series de tiempo en vista de que ambas son de orden de integración 1, es decir que son similares a una recta con pendiente e intercepto. 



```python
# deteccion de los dias que tienen más de una observación y reducimos estas observaciones a una sola utilizando su media 
x = serie.groupby(["fecha"])["variable1"].count()
x = pd.DataFrame(x)
x = x.reset_index()
x[x.variable1 > 1]   # datos de los días en donde se ´presenta más de una observacion 
```

```python
# limpieza y rellenado de los datos 
# reduccion de observaciones 
x = serie.groupby(["fecha"])[ "variable1"].mean()
x = pd.DataFrame(x)
x = x.reset_index()
y = serie.groupby(["fecha"])[ "variable2"].mean()
y = pd.DataFrame(y)
y = y.reset_index()


start = x.fecha[0]
end = x.fecha[len(x)-1]
new_dates = pd.date_range(start=start,end=end,freq='D')
new_dates =  pd.DataFrame(new_dates)
new_dates.columns = ['fecha']
new_dates['fecha'] = new_dates['fecha'].dt.strftime('%Y-%m-%d')
new_dates = pd.merge(new_dates, x, how='left')
new_dates = pd.merge(new_dates, y, how='left')

a = new_dates.variable1.isna()
new_dates.loc[ a, :]
new_dates[['variable1']].plot(title='Variable 1 con nulos y rango completo')
new_dates[['variable2']].plot(title='Variable 2 con nulos y rango completo')

plt.show()
```

```python
new_dates = new_dates.ffill()   ##### rellenamos los nulos con la ultima observacion vista antes del nulo 
new_dates[['variable1']].plot(title='Variable 1 sin nulos y rango completo')
new_dates[['variable1']].plot(title='Variable 2 sin nulos y rango completo')

plt.show()
```

# Selección de modelo para predicción



Como he podido experimentar es muy difícil ganarle a la selección del modelo que realiza la implementación de  los modelos que realiza la función `auto_arima` (del lenguaje R) y que tiene un port en Python la utilizó para dejar que seleccione un modelo univariado para cada serie. 

Solo se reportan los resultados finales, en el código se puede distinguir que en el llamado a la función se indican los parámetros `seasonal`=True y `m=7` que indican que estamos trabajando con estacionalidad semanal pues la implementación requiere que el analista indique esto. 


```python
from pmdarima.arima import auto_arima
import numpy as np
TEST_SIZE = int(new_dates.shape[0]/10)
print(TEST_SIZE)
data_v = new_dates.copy()

train, test = data_v[['variable1']].iloc[:-TEST_SIZE], data_v[['variable1']].iloc[-TEST_SIZE:]
x_train, x_test = np.array(range(train.shape[0])), np.array(range(train.shape[0], data_v.shape[0]))
model1 = auto_arima(train, start_p=0, start_q=0, test='adf', max_p=5, max_q=5, m=7, d=None, seasonal=True, 
                   start_P=0, D=None, trace=False, error_action='ignore',  suppress_warnings=True, stepwise=True)
model1.summary()
```

```python
train, test = data_v[['variable2']].iloc[:-TEST_SIZE], data_v[['variable2']].iloc[-TEST_SIZE:]
x_train, x_test = np.array(range(train.shape[0])), np.array(range(train.shape[0], data_v.shape[0]))
model2 = auto_arima(train, start_p=0, start_q=0, test='adf', max_p=5, max_q=5, m=4, d=None, seasonal=True, 
                   start_P=0, D=None, trace=False, error_action='ignore',  suppress_warnings=True, stepwise=True)
model2.summary()
```

```python
prediction = model1.predict(n_periods=TEST_SIZE, return_conf_int=False)
prediction_series = pd.DataFrame(prediction)
prediction_series.columns = ['variable1']
prediction_series.index = new_dates.fecha[  (len(train)): ]
prediction_series= prediction_series.reset_index()
prediction_series.fecha = pd.to_datetime(prediction_series.fecha) 
new_dates.fecha = pd.to_datetime ( new_dates.fecha) 
prediction_series['variable2'] = model2.predict(n_periods=TEST_SIZE, return_conf_int=False).values

new_dates
```

```python
new_dates["variable1"] = (new_dates["variable1"])**1
new_dates["variable2"] = (new_dates["variable2"]) **1
prediction_series["variable1"]= (prediction_series["variable1"])**1
prediction_series["variable2"]= (prediction_series["variable2"])**1
```

# Problema de salvación 





En el 2019 implemente un algoritmo para realizar pronósticos de alta dimensionalidad $m>n$, con muchas variables $m$ y con muchas observaciones $n$. 

El algoritmo utiliza la técnica de PLS para reducir dimensionalidad y los modelos VAR para realizar la modelación. 
Hay dos puntos técnicos importantes en este algoritmo y en su implementación que quiero destacar: 
Para evitar el problema de invertir la matriz de coeficientes de los modelos VAR, el método realiza una prueba de cointegración por pares de variables para mostrar  que esta relación existe en el conjunto de todas las variables. 
Se implementó el subproblema anterior y para hacerlo eficiente se realizó en paralelo utilizando una arquitectura multihilo en el lenguaje R. 

El codigo puede ser consultado [aqui](https://github.com/fou-foo/Tesis/tree/master/Code) y un video con una explicación de como funciona el método (a partir del minuto3  puede consultarse [aca](https://sg.com.mx/buzz/ponencias/data-day-monterrey/pronosticos-mejorados-por-medio-de-var-pls)).
